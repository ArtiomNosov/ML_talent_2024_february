```python
def prepare_resumes(data, resume_type):
    uuid_failed_resumes_list = []
    first_name_list = [] 
    last_name_list = [] 
    birth_date_list = [] 
    country_list = [] 
    city_list = [] 
    about_list = [] 
    key_skills_list = [] 
    experienceItem_employer_list = []
    experienceItem_city_list = [] 
    experienceItem_position_list = [] 
    experienceItem_description_list = [] 
    experienceItem_sum_time_list = [] 

    educationItem_year_list = []
    educationItem_organization_list = []
    educationItem_faculty_list = []
    educationItem_specialty_list = []
    educationItem_result_list = []
    educationItem_education_type_list = []
    educationItem_education_level_list = []
    uuid_vocancy = []
    for i in range(len(data)):
        for j in range(len(data[i][resume_type])):
            uuid_failed_resumes_list.append(data[i][resume_type][j]['uuid'])
            first_name_list.append(data[i][resume_type][j]['first_name']) 
            last_name_list.append(data[i][resume_type][j]['last_name']) 
            birth_date_list.append(data[i][resume_type][j]['birth_date']) 
            country_list.append(data[i][resume_type][j]['country']) 
            city_list.append(data[i][resume_type][j]['city']) 
            about_list.append(data[i][resume_type][j]['about']) 
            key_skills_list.append(data[i][resume_type][j]['key_skills'])
            uuid_vocancy.append(data[i]['vacancy']['uuid'])
            # print(data_for_members[i]['failed_resumes'][j]['experienceItem'])
            str_experienceItem_employer = ''
            str_experienceItem_city = ''
            str_experienceItem_position = ''
            str_experienceItem_description = ''
            sum_experience_time = 0
            if 'experienceItem' in data[i][resume_type][j]:
                for experienceItem in data[i][resume_type][j]['experienceItem']:
                    # print(experienceItem)
                    if experienceItem['ends'] != None:
                        experience_time = datetime.strptime(experienceItem['ends'], '%Y-%m-%d') - datetime.strptime(experienceItem['starts'], '%Y-%m-%d')
                    else:
                        experience_time = datetime.now() - datetime.strptime(experienceItem['starts'], '%Y-%m-%d')
                    if sum_experience_time == 0:
                        sum_experience_time = experience_time
                    else:
                        sum_experience_time += experience_time
                    str_experienceItem_employer += ',' + experienceItem['employer']
                    str_experienceItem_city += ',' + experienceItem['city']
                    str_experienceItem_position += ',' + experienceItem['position']
                    str_experienceItem_description += ',' + str(experienceItem['description'])
                if sum_experience_time != 0:
                    sum_experience_time = sum_experience_time.days
                # print(sum_experience_time/365)
                experienceItem_employer_list.append(str_experienceItem_employer)
                experienceItem_city_list.append(str_experienceItem_city)
                experienceItem_position_list.append(str_experienceItem_position)
                experienceItem_description_list.append(str_experienceItem_description)
                experienceItem_sum_time_list.append(sum_experience_time/365)
            else:
                experienceItem_employer_list.append(None)
                experienceItem_city_list.append(None)
                experienceItem_position_list.append(None)
                experienceItem_description_list.append(None)
                experienceItem_sum_time_list.append(0.0)





            str_educationItem_year = ''
            str_educationItem_organization = ''
            str_educationItem_faculty = ''
            str_educationItem_specialty = ''
            str_educationItem_result = ''
            str_educationItem_education_type = ''
            str_educationItem_education_level = ''
            if 'educationItem' in data[i][resume_type][j]:
                for educationItem in data[i][resume_type][j]['educationItem']:
                    str_educationItem_year += ',' + str(educationItem['year'])
                    str_educationItem_organization += ',' + str(educationItem['organization'])
                    str_educationItem_faculty += ',' + educationItem['faculty']
                    str_educationItem_specialty += ',' + str(educationItem['specialty'])
                    str_educationItem_result += ',' + str(educationItem['result'])
                    str_educationItem_education_type += ',' + educationItem['education_type']
                    str_educationItem_education_level += ',' + str(educationItem['education_level'])

                educationItem_year_list.append(str_educationItem_year)
                educationItem_organization_list.append(str_educationItem_organization)
                educationItem_faculty_list.append(str_educationItem_faculty)
                educationItem_specialty_list.append(str_educationItem_specialty)
                educationItem_result_list.append(str_educationItem_result)
                educationItem_education_type_list.append(str_educationItem_education_type)
                educationItem_education_level_list.append(str_educationItem_education_level)
            else:
                educationItem_year_list.append(None)
                educationItem_organization_list.append(None)
                educationItem_faculty_list.append(None)
                educationItem_specialty_list.append(None)
                educationItem_result_list.append(None)
                educationItem_education_type_list.append(None)
                educationItem_education_level_list.append(None)
            
    return uuid_failed_resumes_list, first_name_list, last_name_list, birth_date_list, country_list, city_list, about_list , key_skills_list , experienceItem_employer_list, experienceItem_city_list , experienceItem_position_list , experienceItem_description_list , experienceItem_sum_time_list , educationItem_year_list, educationItem_organization_list, educationItem_faculty_list, educationItem_specialty_list, educationItem_result_list, educationItem_education_type_list, educationItem_education_level_list, uuid_vocancy

uuid_failed_resumes_list, first_name_list, last_name_list, birth_date_list, country_list, city_list,about_list , key_skills_list , experienceItem_employer_list, experienceItem_city_list , experienceItem_position_list , experienceItem_description_list , experienceItem_sum_time_list , educationItem_year_list, educationItem_organization_list, educationItem_faculty_list, educationItem_specialty_list, educationItem_result_list, educationItem_education_type_list,educationItem_education_level_list, uuid_v_r_list =  prepare_resumes(data_for_members, 'failed_resumes')
uuid_failed_resumes_list_c, first_name_list_c, last_name_list_c, birth_date_list_c, country_list_c, city_list_c, about_list_c, key_skills_list_c , experienceItem_employer_list_c, experienceItem_city_list_c , experienceItem_position_list_c , experienceItem_description_list_c , experienceItem_sum_time_list_c , educationItem_year_list_c, educationItem_organization_list_c, educationItem_faculty_list_c, educationItem_specialty_list_c, educationItem_result_list_c, educationItem_education_type_list_c,educationItem_education_level_list_c, uuid_v_r_list_c =  prepare_resumes(data_for_members, 'confirmed_resumes')
```
